### 2.3. Refining the analysis
#### Apart from client specific queries, the initial data set also provides information about the city's general infrastructure with regard to its dog population. Thus for example, we can explore relationships between the densitiy of amentities and services. One obvious question would be if the availibility of recreation space is somehow compromised by a high demand. Unfortunately we do not have access to information about the number of dogs living in each of the city's boroughs, but we can use a cummulative score as indicator. Let's assume that veterinarians and petshops locate in areas where many pets live, and add to it the density of dog trainers. This cummulative indicator can be compared to the number of recreation grounds (on and off leash combined) in each borough to get an idea about the availibility of dog walking space in relation to the dog population density.

```python
#creating an indicator for dog population density
##selecting amenities/services
df = df_data_0
df_data_3 = df.loc(axis=1)['Vets', 'Petshops', 'Training']
result = df_data_3
print(result)
```

```python
##calculating total score
df_data_3b=df_data_3.transpose()
df_data_3b.sum(axis=None, skipna=0, level=None, numeric_only=None, min_count=0)
```

```python
##sorting and printing the first five boroughs
df3=df_data_3b.sum(axis=None, level=None, numeric_only=None, min_count=0)
result = df3
df3.sort_values(inplace=True, ascending=False)
df3.head()
```

```python
#creating cummulative score for on and off leash recreational grounds
df = df_data_0
df_data_4 = df.loc(axis=1)['RG on leash', 'RG off leash']
result = df_data_4
print(result)
```

```python
##calculating total score
df_data_4b=df_data_4.transpose()
df_data_4b.sum(axis=None, skipna=0, level=None, numeric_only=None, min_count=0)
```

```python
##sorting and printing the first five boroughs
df4=df_data_4b.sum(axis=None, level=None, numeric_only=None, min_count=0)
result = df4
df4.sort_values(inplace=True, ascending=False)
df4.head()
```

```python
cdf3 = df3
cdf4 = df4
plt.scatter(cdf4, cdf3,  color='blue')
plt.xlabel("Recreation Grounds")
plt.ylabel("Dog Population Density")
plt.show()
```

```python
from sklearn import linear_model
regr = linear_model.LinearRegression()

y=np.array([19, 14, 13, 13, 13, 12, 11, 11, 10, 10, 10, 10, 10, 10, 9, 9, 9, 9, 8, 8, 8, 8, 8, 6, 5])
x=np.array([13, 12, 11, 10, 10, 10, 9, 9, 9, 9, 9, 8, 8, 8, 7, 7, 7, 6, 6, 6, 6, 6, 6, 5, 5])
plt.plot(x, y, 'o')
m, b = np.polyfit(x, y, 1)
np.polyfit(x, y, 1)
plt.plot(x, m*x + b)
plt.xlabel("Recreation Grounds")
plt.ylabel("Dog Population Density")
```

#### The linear regression model demonstrates a high correlation between dog population density and availability of recreation grounds, suggesting that supply and demand are fairly balanced.

## Conclusion

#### The initial quest was to develop a model that would allow dog-owners to determine the dog-friendliness of Munich's 25 boroughs. Typical needs such as training facilities, green areas, and veterinarians were identified and the relevant data was gathered from public sources. The data was loaded into a pandas dataframe so that any selection of values and combinations of selections would produce a score and ranking by a simple sort algorithm. The data was visualized in a bar diagramme and choropleth map to allow for a comparison of geographical conditions and individual queries. Finally, the data set was explored by means of a simple linear regression model to test the correlation between two variables derived from the data set: does the availability of recreation grounds (independent variable) impact on the dog pupolation density (dependent variable)? Visualizing the linear regression of the variables in a scatter plot demonstrates a high correlation between the number of dogs and recreation space in each borough, hence we can assume an even geographical distribution accross Munich. The linear regression model can be used to improve the reliabilty of the ranking by analysing the relationship between mere administrative boundaries and actual geographic conditions and potential bias deriving thereof.
